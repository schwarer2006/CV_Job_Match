```python
import matplotlib.pyplot as plt
import numpy as np

# Kategorien definieren
labels = np.array([
    'SQL Server',
    'ETL Design',
    'OLAP/Tabular Design',
    'Reporting',
    'T-SQL/MDX Programming',
    'Python',
    'Professional Experience',
    'Independent & Team Work',
    'German & Communication'
])

# Meine Kenntnisse in Prozent
my_stats = np.array([90, 90, 80, 90, 90, 50, 100, 100, 100])

# Stellenprofilanforderungen in Prozent
job_req_stats = np.array([100, 100, 100, 100, 100, 75, 90, 90, 100])

# Anzahl der Variablen
num_vars = len(labels)

# Winkel berechnen, an denen die Labels platziert werden
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Der Radar-Chart ist ein Kreis, also muss der Array vervollst채ndigt werden
my_stats = np.concatenate((my_stats,[my_stats[0]]))
job_req_stats = np.concatenate((job_req_stats,[job_req_stats[0]]))
angles += angles[:1]

# Die Figur zeichnen
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles, my_stats, color='blue', alpha=0.25)
ax.plot(angles, my_stats, color='blue', linewidth=2, label='Meine Kenntnisse')  # Linien f체r meine Kenntnisse zeichnen
ax.fill(angles, job_req_stats, color='red', alpha=0.25)
ax.plot(angles, job_req_stats, color='red', linewidth=2, label='Stellenprofilanforderungen')  # Linien f체r Stellenanforderungen zeichnen

# Labels platzieren
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Legende hinzuf체gen
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Titel setzen
plt.title('Vergleich der Kompetenzen mit Stellenanforderungen')

plt.show()
```
